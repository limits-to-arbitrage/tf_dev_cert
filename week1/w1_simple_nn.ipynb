```python
from tensorflow import keras
import tensorflow as tf
import numpy as np
```

```python
# The 'hello world' of TensorFlow
# Make and compile simple neural network
model = tf.keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
model.compile(optimizer='sgd', loss='mean_squared_error')
```

```python
# Make input data
xs = np.array(range(-1, 5), dtype=float)
ys = 2*xs - 1
```

```python
# Train
model.fit(x=xs, y=ys, epochs=1_000)
```

```python
# Make prediction with unseen x value
prediction = model.predict([10])
print(f'\nModel predicted {prediction[0][0]} versus expectation of 19.')
```
